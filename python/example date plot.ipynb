```python
import datetime as dt
from datetime import timedelta
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates

# producing example data
n = 20
x = sorted([dt.datetime.now() + timedelta(hours = i) for i in range(n)])
y = np.random.random(n)
```

```python
# normal plot
plt.figure(figsize=(15, 8))
plt.plot(x, y)
```

```python
# see also https://www.programcreek.com/python/example/61484/matplotlib.dates.HourLocator
# with datelocator
fig = plt.figure(figsize=(25, 8), dpi=120)
#fig, ax = plt.subplots(1,1,figsize(25,8))
ax = plt.subplot(1,1,1)
plt.plot(x, y)

ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter("%d.%m.")) 

ax.xaxis.set_minor_locator(mdates.HourLocator(byhour = range(24)))
ax.xaxis.set_minor_formatter(mdates.DateFormatter("%H:%M"))

fig.autofmt_xdate()
```
